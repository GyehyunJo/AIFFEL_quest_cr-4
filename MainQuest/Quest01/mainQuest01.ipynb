# 메인 퀘스트 1번: 은행계좌 만들기

```python
# random함수를 쓰기위해 import
import random as rd

# Q1. Account 클래스 생성
# 은행이름, 예금주, 계좌번호, 잔액이 필요. 생성 될때 예금주와 초기 잔액만 입력 받음.
# 은행이름은 SC은행, 계좌번호는 3자리-2자리-6자리 형태로 랜덤 생성.
class Account:
    # Q2. 클래스 변수: 생성된 계좌 객체 수 저장.
    account_count = 0

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank = "SC은행"

        # 3자리 2자리 6자리로 정수를 나눠서 받을 것임.
        num1 = rd.randint(0, 999)
        num2 = rd.randint(0, 99)
        num3 = rd.randint(0, 999999)

        # 랜덤으로 나온 값에 빈자리를 0으로 채움. ex: 312 >> 312, 12 >> 012
        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)
        # 세부분을 하나로 합침.
        self.account_number = num1 + '-' + num2 + '-' + num3

        Account.account_count += 1

        # 이자를 구하기 위한 입금 횟수 카운트, 최종 횟수 카운트.
        self.deposit_count = 0
        self.count = 0
        # 입금, 출금 따로 저장, 최종 순서 저장.
        self.deposit_log = []
        self.withdraw_log = []
        self.history_log = []

    # Q3. 클래스 변수: 생성된 총 계좌수 출력
    @classmethod
    def get_account_num(cls):
        return (cls.account_count)

    # Q4. 입금 메서드: 입금을 위한 deposit 메서드 추가. 최소 1원 이상부터 가능.
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.count += 1
            # 입금 내역 저장.
            self.deposit_log.append(amount)
            self.history_log.append(f'입금\t금액: {amount}\t잔액: {self.balance}')
            # Q7. 이자 지급하기: 입금 횟수가 5회 마다 잔고의 1% 이자 추가.
            if self.deposit_count % 5 == 0:
                self.interest = int(self.balance * 0.01)
                print(f'{self.interest}의 이자가 입금되었습니다.')
                self.balance += self.interest
                self.count += 1
                # 이자 지급 내역 저장.
                self.history_log.append(f'이자지급\t금액: {self.interest}\t잔액: {self.balance}')
        else:
            print("최소 1원 이상 입금이 가능합니다.")
    # Q5. 출금 메서드: 출급을 위한 withdraw 메서드 추가. 계좌 잔고 이상으로 출금 불가능.
    def withdraw(self, amount):
        if self.balance > amount:
            self.balance -= amount
            self.count += 1
            # 출금 내역 저장.
            self.withdraw_log.append(amount)
            self.history_log.append(f'출금\t금액: {amount}\t잔액: {self.balance}')
        else:
            print("잔액부족으로 출금할 수 없습니다.")

    # Q6.정보 출력 메서드: Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드 추가. 잔고는 세자리마다 쉼표 출력.
    def display_info(self):
        print(f'은행 이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}') # {self.balance:,}

    # Q10. 입금과 출금 내역이 기록하라. 각각 deposit_history와 withsraw_history 메서드 추가.
    def deposit_history(self):
        for log in self.history_log:
            if '입금' in log or '이자지급' in log:
                print(f'{self.history_log.index(log)+1}회: {log}')
            else:
                pass

    def withdraw_history(self):
        for log in self.history_log:
            if '출금' in log:
                print(f'{self.history_log.index(log)+1}회: {log}')
            else:
                pass

# Q8. 여러 객체 생성: Account 클래스로부터 3개 이상 인스턴스를 생성하고 리스트에 저장하기.
bank_list = []
'''
원레 예시로 쓸려고 했던 코드.

kang = Account("강민수", 100)
lee = Account("이민수", 1000000)
park = Account("박민수", 985000)

bank_list.append(kang)
bank_list.append(lee)
bank_list.append(park)

# Q9. 객체 순회 반복문을 톨해 리스트에서 잔고가 100만원 이상인 고객의 정보 출력하기.
for account in bank_list:
    if account.balance >= 1000000:
        account.display_info()
print()

# Q7, Q10 출력 확인
kang.deposit(100)
kang.deposit(100)
kang.deposit(100)
kang.deposit(50)
kang.deposit(50)
# 입금 로그
kang.deposit_history()
print()

# 이자 확인
kang.display_info()
print()

kang.withdraw(100)
# 출금 로그
kang.withdraw_history()
print()
'''

frist_account = Account('차정은', 1000)
second_account = Account('박광석', 30000000)
third_account = Account('조웅제', 50000000)

bank_list.append(frist_account)
bank_list.append(second_account)
bank_list.append(third_account)

acc_num = Account.get_account_num()
print('생성된 계좌의 총 개수: ',acc_num)

for account in bank_list:
    if account.balance >= 1000000:
        account.display_info()

print()

my_account = Account('차정은', 1000)

for i in range(10):
    my_account.deposit(1000)
my_account.deposit(-1000)
my_account.withdraw(3000000)
my_account.withdraw(3000)

print()

my_account.deposit_history()
print('-' * 50)
my_account.withdraw_history()
```

# 메인 퀘스트 2번: 간단한 자동사냥 RPG 만들기

간단한 자동사냥 RPG 게임을 만들자! 사용자의 이름을 입력 받아 플래이어를 생성하고, 몬스터들을 차례로 자동사냥하는 게임을 만든다.

```python
# random을 쓰기위한 import
import random as rd

# Q1. Character 클래스 만들기
# 이름, 레벨, 체력, 공격력, 방어력의 속성을 가짐.
# 인스턴스의 현재 체력이 0 이상인지 bool 값을 반환하는 is_alive 메서드 만들기.
# 공격을 받았을 때, (받은 데미지 - 본인의 방어력)만큼 현재 체력이 감소하는 take_damage 메서드 만들기. 방어력이 더 높을 시 체력 감소 없음.
# 타겟에게 데미지를 입히는 attack_target 메서드 만들기. 데미지는 1부터 공격력 사이의 램덤 정수.
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        if self.hp > 0:
            return True
        else:
            return False

    def take_damage(self, damage):
        if damage > self.defense:
            self.hp -= (damage - self.defense)
        else:
            pass

    def attack_target(self, target):
        damage = rd.randint(1, self.attack)
        target.take_damage(damage)
        print(f'{self.name}이 {target.name} 에게 {damage} 만큼 공격했다...!')
        print(f'{target.name}의 체력: {target.hp}')

# Q2. Player 클래스와 Monster 클래스 만들기.
# 두 클래스 모두 Character 클래스를 상속 받음.

# Player 클래스
# Character 클래스를 상속 받기.
# 레벨 1, 체력 100, 공격력25, 방어력 5로 초기화
# Player 클래스는 경험치 속성을 추가로 가짐.
# 인수로 받은 정수 만큼 경험치를 획득하는 gain_experience 메서드 만들기.
# 현재 경험치가 50이상이면 레벨을 1, 공격력을 10, 방어력을 5씩 올리는 level_up 메서드 만들기.
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp

        if self.experience >= 50:
            print(f'{exp}경험치를 얻었다.\n')
            self.level_up()
            print(f'\n{self.name}의 다음 레벨까지 필요 경험치: {50 - self.experience}')
        else:
            print(f'{exp}경험치를 얻었다.\n{self.name}의 다음 레벨까지 필요 경험치: {50 - self.experience}')

    def level_up(self):
        self.level += 1
        self.attack += 10
        self.defense += 5
        self.experience -= 50
        print(f'레벨 업!\n{self.name}의 레벨이 {self.level} 이(가) 되었습니다.\n')
        print('◤▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬◥')
        print(f' {self.name}의 레벨: {self.level} (+1)\n 체력\t: {self.hp}\n 공격력\t: {self.attack} (+10)\n 방어력\t: {self.defense} (+5)')
        print('◣▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬▬◢')

# Monster 클래스
# Character 클래스를 상속 받기.
# 몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화하기.
# - 체력  : 10 ~ 30 사이의 랜덤한 정수 * 레벨
# - 공격력: 5 ~ 20 사이의 랜덤한 정수 * 레벨
# - 방어력: 1 ~ 5 사이의 램덤한 정수 * 레벨
class Monster(Character):
    def __init__(self, name, level):
        self.hp = rd.randint(10, 30) * level
        self.attack = rd.randint(5, 20) * level
        self.defense = rd.randint(1, 5) * level
        super().__init__(name, level, self.hp, self.attack, self.defense)

# Q3. battle 함수 만들기
# Player 인스턴스와 Monster 인스턴스를 인수로 받아 둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고 받는 함수.
# 만약 Player 인스턴스가 살아남으면 >> Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가, player의 레벨업 메서드 호출, '전투 승리!' 출력
# 만약 Monster 인스턴스가 살아남으면 >> '전투 패배...' 출력
def battle(player, monster):
    print(f'{monster.name} 과(와)의 전투를 시작합니다.')
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
    if player.is_alive():
        print('\n전투 승리!')
        print(f'{monster.name} 을(를) 이겼다!\n')
        player.gain_experience(monster.level * 20)
        print('-' * 50)
    else:
        print('\n전투 패배...')
        print('-' * 50)

# Q4. main 함수 만들기
# 몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기. ex: monster_dict = {'슬라임':1,'고블린':2,'오크':3}
# 사용자로부터 이름을 받아 Player 인스턴스 생성.
# 몬스터 딕셔너리로부터 Monster 인스턴스 todtjd.
# 생성된 Monster 인스턴스와 Player 인스턴스가 battle 함수를 통해 전투
# Player는 생선된 몬스터 3마리(슬라임, 고블린, 오크)와 모두 전투 해야함.
# 전투 도중 Player가 사망하면 이후 전투는 진행하지 않고 '게임오버'출력 후 종료.
def main():
    monster_dict = {'슬라임':1,'고블린':2,'오크':3}

    player_name = input('플레이어의 이름을 입력하세요: ')
    player = Player(player_name)

    print('-' * 50)
    print(f'플레이어 이름: {player.name}\n레벨: {player.level}\n체력: {player.hp}\n공격력: {player.attack}\n방어력: {player.defense}')
    print('-' * 50)

    for name, level in monster_dict.items():
        monster = Monster(name, level)
        battle(player, monster)

        if not player.is_alive():
            print('게임오버')
            break

main()
```
